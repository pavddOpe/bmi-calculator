## BMI Calculator 
*using SI - International System of Units measurement*

```python
from IPython.display import HTML, display

display(HTML("""
<style>
body {
    background-color: #1956d0;
    font-family: system-ui, -apple-system, BlinkMacSystemFont, sans-serif;
}

.app {
    max-width: 420px;
    margin: 60px auto;
    padding: 30px;
    background: white;
    border-radius: 16px;
    box-shadow: 0 10px 30px rgba(0,0,0,0.08);
}

button {
    width: 100%;
    margin-top: 10px;
}
</style>
"""))
```

```python
import ipywidgets as widgets
from IPython.display import Markdown, display


output = widgets.Output()

title = widgets.HTML("<h2 style='text-align:center;'>BMI Calculator</h2>")

name_input = widgets.Text(
    description="Name:",
    placeholder="Enter your name"
)

weight_input = widgets.FloatText(
    description="Weight (kg):"
)

height_input = widgets.FloatText(
    description="Height (m):"
)

button = widgets.Button(
    description="Calculate BMI",
    button_style="primary"
)


def calc_bmi(_):
    with output:
        output.clear_output()

        name = name_input.value
        weight = weight_input.value
        height = height_input.value

        if weight <= 0 or height <= 0:
            display(Markdown("❌ **Please provide valid measurements.**"))
            return

        bmi = weight / (height ** 2)

        if bmi < 18.5:
            status = "Underweight"
        elif bmi <= 24.9:
            status = "Normal weight"
        elif bmi <= 29.9:
            status = "Overweight"
        elif bmi <= 34.9:
            status = "Obese"
        elif bmi <= 39.9:
            status = "Severely obese"
        else:
            status = "Morbidly obese"

        display(Markdown(
            f"""
### ✅ Result
**Name:** {name}  
**BMI:** `{bmi:.2f}`  
**Status:** **{status}**
"""
        ))


button.on_click(calc_bmi)

app = widgets.VBox(
    [
        title,
        name_input,
        weight_input,
        height_input,
        button,
        output
    ],
    layout=widgets.Layout(
        padding="20px"
    )
)

display(widgets.HTML("<div class='app'>"), app, widgets.HTML("</div>"))
```
